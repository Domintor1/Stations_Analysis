```python
#Renaming stations total
df.rename(columns={'Station Total':'Stations Total'}, inplace = True)
Stations_Total = df['Stations Total']
```

```python
# Pivot Tables
```

```python
peak_pt = df.pivot_table(index= 'Peak', values = ['Waterloo', 'Victoria', 'Charing Cross', 'Cannon Street', 'Claphan Junction'], aggfunc= 'sum')
```

```python
# Pivot table of stations to analyse in particular for visualisations
peak_pt[['Waterloo', 'Victoria', 'Charing Cross', 'Cannon Street', 'Claphan Junction']]
```

```python
months_pt = df.pivot_table(index='Log Time',values= ['Waterloo', 'Victoria', 'Charing Cross', 'Cannon Street', 'Claphan Junction'], aggfunc='sum')
```

```python
months_pt
```

```python
day_pt = df.pivot_table(index='Date', values = ['Waterloo', 'Victoria', 'Charing Cross', 'Cannon Street', 'Claphan Junction'], aggfunc='sum')
day_pt
```

```python
# Visualisations
peak_pt.plot(kind='bar', title='Visitors per Station According To Peak Time', xlabel= 'Time of Day', ylabel='Visitors in Millions')
```

```python
peak_pt.plot.pie(y='Cannon Street', figsize=(10,10))
```

```python
months_pt.plot(kind='line', xlabel='Time of Day', ylabel='Number of Visitors', title='Number of Visitors By Time Of Day')
```
